```python
freq_range = (0, 1, 100)
hht_f0, spec0 = emd.spectra.hilberthuang(IF[:,0], IA[:,0],freq_range)
hht_f1, spec1 = emd.spectra.hilberthuang(IF[:,1], IA[:,1],freq_range)
hht_f2, spec2 = emd.spectra.hilberthuang(IF[:,2], IA[:,2],freq_range)

hht_f, spec = emd.spectra.hilberthuang(IF, IA,freq_range)


plt.plot(hht_f0, spec0)
plt.plot(hht_f1, spec1)
plt.plot(hht_f2, spec2)
plt.plot(hht_f, spec)
plt.semilogy()
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'nht')
plt.figure(figsize=(8, 4))

plt.subplot(121)
# Plot a simple histogram using frequency bins from 0-.5Hz
plt.hist(IF[:, 1], np.linspace(0, .5,30))
plt.grid(True)
plt.title('IF Histogram')
plt.xlabel('Frequency (Hz)')

plt.subplot(122)
# Plot an amplitude-weighted histogram using frequency bins from 0-.5Hz
plt.hist(IF[:, 1], np.linspace(0, .5,30), weights=IA[:, 1])
plt.grid(True)
plt.title('IF Histogram\nweighted by IA')
plt.xlabel('Frequency (Hz)')
```

# Hilbert Spectrogram

### Normal EMD

```python
IP, IF, IA = emd.spectra.frequency_transform(imf[:,1], fs, 'hilbert')
freq_range = (0, .6, 100)
hht_f, hht = emd.spectra.hilberthuang(IF, IA, freq_range, mode='amplitude', sum_time=False)
hht = ndimage.gaussian_filter(hht, 1.4)
plt.figure(figsize=(10,5))

c=plt.pcolormesh(t1, hht_f, hht, cmap='Blues')
plt.title('Hilbert-Huang Transform')
plt.xlabel('Time (mi)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(c)
```

### Masked EMD

```python
imfm = emd.sift.mask_sift(x)

IPm, IFm, IAm = emd.spectra.frequency_transform(imfm[:,:3], fs, 'hilbert')
freq_range = (0, .5, 100)
hht_f, hht = emd.spectra.hilberthuang(IFm, IAm, freq_range, mode='amplitude', sum_time=False)
hht = ndimage.gaussian_filter(hht, 1.5)
plt.figure(figsize=(10,5))

c= plt.pcolormesh(t1, hht_f, hht, cmap='Blues')
plt.title('Hilbert-Huang Transform')
plt.xlabel('Time (min)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(c)
```

## Relevance and Quality of IMFs

```python

def quality(c_i):
    a,b,c=0,0,0
    for n in range(len(t1)):
        a+= x[n]*c_i[n]
        b+= x[n]**2
        c+= c_i[n]**2

    gamma= a/(np.sqrt(b*c))
    return gamma
```

```python
for i in range(4):
    print(quality(imf[:,i]))

print(quality(imf[:,0]+imf[:,1]+imf[:,2]+imf[:,3]))
```
