### quality

```python
def quality(c_i):
    a,b,c=0,0,0
    for n in range(len(t1)):
        a+= x[n]*c_i[n]
        b+= x[n]**2
        c+= c_i[n]**2

    gamma= a/(np.sqrt(b*c))
    return gamma
```

```python
for i in range(3):
    print(quality(imf[:,i]))
print(quality(imf[:,1]+imf[:,2]+imf[:,0]))
```
