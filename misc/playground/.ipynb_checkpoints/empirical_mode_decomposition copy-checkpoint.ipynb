### EMD

```python
x=np.array(wp1)
```

```python
imf = emd.sift.sift(x)
```

```python
# Get the default configuration for a sift
config = emd.sift.get_config('sift')
# Adjust the threshold for accepting an IMF
config['imf_opts/sd_thresh'] = 0.05
config['extrema_opts/method'] = 'rilling'

imf = emd.sift.sift(x)

# emd.plotting.plot_imfs(imf)
```

```python
# Extract the options for get_next_imf - these can be customised here at this point.
imf_opts = config['imf_opts']

#Extract first IMF from the signal
imf1, continue_sift = emd.sift.get_next_imf(x[:, None],           **imf_opts)

# Extract second IMF from the signal with the first IMF removed
imf2, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1,      **imf_opts)

# Extract third IMF from the signal with the first and second IMFs removed
imf3, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2, **imf_opts)

imf4, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2-imf3, **imf_opts)

imf5, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2-imf3-imf4, **imf_opts)


# The residual is the signal component left after removing the IMFs
residual = x[:, None] - imf1 - imf2 - imf3-imf4-imf5

#Contactenate our IMFs into one array
imfs_manual = np.c_[imf1, imf2, imf3,imf4,imf5, residual]

# Visualise
# emd.plotting.plot_imfs(imfs_manual)
```

```python
def list_corrector(l):
    l_=[]
    for i in range(len(l)):
        l_.append(l[i][0])
    return np.array(l_)
```

```python
imf1=list_corrector(imf1)
imf2=list_corrector(imf2)
imf3=list_corrector(imf3)
imf4=list_corrector(imf4)
```

### Hilbert Transform

```python
# Compute frequency statistics
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'hilbert')
```

```python
plt.plot(t1,IF)
```

```python
IF_1=[]
for i in IF:
    IF_1.append(i[0])
IF_2=[]
for i in IF:
    IF_2.append(i[1])
IF_3=[]
for i in IF:
    IF_3.append(i[2])
IF_4=[]
for i in IF:
    IF_4.append(i[3])
```

```python
print(mean(IF_4))
print(mean(IF_3))
print(mean(IF_2))
print(mean(IF_1))
```

```python
print((mean(IF_4)+mean(IF_3)+mean(IF_2)+mean(IF_1))/4)
```

```python
freq_range = (0, 1, 50)
hht_f, spec = emd.spectra.hilberthuang(IF, IA,freq_range)


plt.plot(hht_f, spec)
plt.legend('Hilbert-Huang Transform', frameon=False)
plt.xlim(0.05, .8)
plt.ylim(0, 2)

plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'nht')
plt.figure(figsize=(8, 4))

plt.subplot(121)
# Plot a simple histogram using frequency bins from 0-20Hz
plt.hist(IF[:, 1], np.linspace(0, .5))
plt.grid(True)
plt.title('IF Histogram')
plt.xlabel('Frequency (Hz)')

plt.subplot(122)
# Plot an amplitude-weighted histogram using frequency bins from 0-20Hz
plt.hist(IF[:, 1], np.linspace(0, .5), weights=IA[:, 1])
plt.grid(True)
plt.title('IF Histogram\nweighted by IA')
plt.xlabel('Frequency (Hz)')
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'hilbert')
freq_range = (0, 1, 100)
hht_f, hht = emd.spectra.hilberthuang(IF, IA, freq_range, mode='amplitude', sum_time=False)
hht = ndimage.gaussian_filter(hht, .5)
plt.pcolormesh(t1, hht_f, hht, cmap='viridis')
plt.title('Hilbert-Huang Transform')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
```

```python
imfm = emd.sift.mask_sift(x, max_imfs=6)
freq_range = (0, 1, 100)
hht = ndimage.gaussian_filter(hht, 5)
IPm,IFm,IAm = emd.spectra.frequency_transform(imfm, fs, 'hilbert')
f, hht = emd.spectra.hilberthuang(IFm, IAm, freq_range, mode='amplitude',sum_time=False)
hht = ndimage.gaussian_filter(hht, .5)

plt.pcolormesh(t1, f, hht, cmap='viridis')
```
