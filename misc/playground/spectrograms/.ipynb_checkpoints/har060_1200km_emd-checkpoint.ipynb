```python
name= 'inttime_har060_1200km'
```

```python
df=pd.read_csv('../../datasets/'+name+'.csv')
t1=df['timestamps']
wp1=df['ls']
x=np.array(wp1)

dt= t1[1]-t1[0]
fs= 1/dt
```

```python

def quality(c_i):
    a,b,c=0,0,0
    for n in range(len(t1)):
        a+= x[n]*c_i[n]
        b+= x[n]**2
        c+= c_i[n]**2

    gamma= a/(np.sqrt(b*c))
    return gamma
```

```python
for i in range(4):
    
    fxx, pxx =sp.signal.welch(imfm[:,i],fs,nperseg=len(t1))
    plt.step(1/fxx,pxx)
plt.axvline(x=3.0,linestyle=(0,(5,5)),alpha=.7,c='fuchsia',linewidth=3.5)
plt.loglog()
plt.show()
```

```python
def fourier(sig,sr):
    X = np.fft.fft(sig)
    N = len(X)
    n = np.arange(N)
    T = N/sr
    freq = n/T 
    return freq[:N//2], np.abs(X)[:N//2]
```

```python
plt.figure(figsize = (12, 6))

plt.subplot(121)
L= fourier(imfm[:,0],fs)
plt.step(1/L[0], L[1], 'b')
L= fourier(imfm[:,1],fs)
plt.step(1/L[0], L[1], 'lightblue')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()

plt.subplot(122)
fxx, pxx =sp.signal.welch(imfm[:,0],fs,nperseg=len(t1))
plt.step(1/fxx,pxx,'b')
fxx, pxx =sp.signal.welch(imfm[:,1],fs,nperseg=len(t1))
plt.step(1/fxx,pxx,'lightblue')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()
```
