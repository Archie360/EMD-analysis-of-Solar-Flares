```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import emd
```

```python
name= 'inttime_doppler_1200km'
```

```python
df=pd.read_csv('../../datasets/'+name+'.csv')
t1=df['timestamps']
wp1=df['ls']
x=np.array(wp1)

dt= t1[1]-t1[0]
fs= 1/dt
```

```python
x= x-mean(x)
x=x/np.sqrt(mean(x**2))
```

```python
plt.figure(figsize=(10,4))
plt.plot(t1,x, '-', color='brown', lw=2, label='signl')
plt.xlabel('time [t]')
plt.ylabel('signal')
```

```python
imf = emd.sift.sift(x)
```

```python
plt.figure(figsize=(10,4))
for i in range(len(imf[0,:])):
    plt.plot(t1,imf[:,i], '-', color='brown', lw=2,alpha= 1/(i+1))
# plt.plot(t1,x, '-', color='green', lw=3, label='imf 1')

plt.xlabel('time [t]')
plt.ylabel('signal')
```

```python
# Compute frequency statistics
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'hilbert')
```

```python
freq_range = (0, 1, 50)
plt.figure(figsize=(12, 4))

hht_f0, spec0 = emd.spectra.hilberthuang(IF[:,0], IA[:,0],freq_range)
hht_f1, spec1 = emd.spectra.hilberthuang(IF[:,1], IA[:,1],freq_range)
hht_f2, spec2 = emd.spectra.hilberthuang(IF[:,2], IA[:,2],freq_range)

hht_f, spec = emd.spectra.hilberthuang(IF, IA,freq_range)

plt.subplot(121)
plt.plot(1/hht_f0, spec0)
plt.plot(1/hht_f1, spec1)
plt.plot(1/hht_f2, spec2)
plt.plot(1/hht_f, spec)
plt.semilogy()
# plt.semilogx()
plt.xlim(0,10)

# plt.legend(['1','2','3','4'], frameon=False)


plt.subplot(122)
fig=px.line(y=spec2,x=hht_f2,template="presentation")
plt.plot(1/hht_f2,spec2)
plt.plot(1/hht_f,spec)
# plt.semilogx()
plt.semilogy()
plt.xlim(0,10)
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'nht')
plt.figure(figsize=(8, 6))

# Plot a simple histogram using frequency bins from 0-.5Hz
plt.hist(IF[:, 1], np.linspace(0, .5,40),color='orange',alpha= 1)
plt.hist(IF[:, 1], np.linspace(0, .5,40), weights=IA[:, 1],color='red',alpha=1)

plt.grid(True)
plt.title('IF Histogram : Weighted vs Unweighted')
plt.xlabel('Frequency (Hz)')
plt.legend(['unweighted','weighted'])
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imf, fs, 'hilbert')
freq_range = (0, .6, 100)
hht_f, hht = emd.spectra.hilberthuang(IF[:,:2], IA[:,:2], freq_range, mode='amplitude', sum_time=False)
hht = ndimage.gaussian_filter(hht, 4)
plt.figure(figsize=(10,5))

c=plt.pcolormesh(t1, hht_f, hht, cmap='seismic')
plt.title('Hilbert-Huang Transform')
plt.xlabel('Time (mi)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(c)
```

```python
imfm = emd.sift.mask_sift(x)

IPm, IFm, IAm = emd.spectra.frequency_transform(imfm, fs, 'hilbert')
freq_range = (0, .7, 100)
hht_f, hht = emd.spectra.hilberthuang(IFm[:,:2], IAm[:,:2], freq_range, mode='amplitude', sum_time=False)
hht = ndimage.gaussian_filter(hht, 4)
plt.figure(figsize=(10,5))

c= plt.pcolormesh(t1, hht_f, hht, cmap='seismic')
plt.title('Hilbert-Huang Transform')
plt.xlabel('Time (min)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(c)
```

```python
def quality(c_i):
    a,b,c=0,0,0
    for n in range(len(t1)):
        a+= x[n]*c_i[n]
        b+= x[n]**2
        c+= c_i[n]**2

    gamma= a/(np.sqrt(b*c))
    return gamma
```

```python
def bm(List):
    mins=[]
    maxs=[]
    for n, i in enumerate(List[1:-1]):
            if List[n] < List[n-1] and List[n] < List[n+1]:
                mins.append(List[n])
            elif List[n] > List[n-1] and i> List[n+1]:
                maxs.append(List[n])
    return len(mins)+len(maxs)
```

```python
def fourier(sig,sr):
    X = np.fft.fft(sig)
    N = len(X)
    n = np.arange(N)
    T = N/sr
    freq = n/T 
    return freq[:N//2], np.abs(X)[:N//2]
```

```python

plt.figure(figsize = (12, 6))

plt.subplot(121)
L= fourier(imf[:,0],fs)
plt.step(1/L[0], L[1], 'b')
L= fourier(imf[:,1],fs)
plt.step(1/L[0], L[1], 'lightblue')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()

plt.subplot(122)
fxx, pxx =sp.signal.welch(imfm[:,0],fs,nperseg=len(t1))
plt.step(1/fxx,pxx,'b')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()
```
