```python
df=pd.read_csv('../../datasets/inttime_har020_1200km.csv')
t1=df['timestamps']
wp1=df['ls']
x=np.array(wp1)

dt= t1[1]-t1[0]
fs= 1/dt
```

```python
x=np.array(wp1)
```

```python
imf = emd.sift.sift(x)
```

```python
imfm = emd.sift.mask_sift(x)
```

```python
# Compute frequency statistics
IP, IF, IA = emd.spectra.frequency_transform(imfm, fs, 'hilbert')
```

```python
freq_range = (0, 1, 100)
hht_f, spec = emd.spectra.hilberthuang(IF, IA,freq_range)
plt.step(1/hht_f,spec)
plt.loglog()
```

```python
for i in range(4):
    
    fxx, pxx =sp.signal.welch(imfm[:,i],fs,nperseg=len(t1))
    plt.step(1/fxx,pxx)
plt.axvline(x=3.0,linestyle=(0,(5,5)),alpha=.7,c='fuchsia',linewidth=3.5)
plt.loglog()
plt.ylim(1e1)
plt.show()
```

```python
IP, IF, IA = emd.spectra.frequency_transform(imfm, fs, 'nht')
plt.figure(figsize=(16, 4))


plt.subplot(121)
# Plot a simple histogram using frequency bins from 0-.5Hz
plt.hist(1/IF[:, 1], np.linspace(3, 7,100))
plt.title('IF Histogram')
plt.xlabel(' (Hz)')
plt.semilogy()

plt.subplot(122)
# Plot an amplitude-weighted histogram using frequency bins from 0-.5Hz
plt.hist(1/IF[:, 1], np.linspace(3, 7, 100), weights=IA[:, 1])
plt.title('IF hist weighted by IA')
plt.xlabel(' (Hz)')
plt.semilogy()
```

```python

def quality(c_i):
    a,b,c=0,0,0
    for n in range(len(t1)):
        a+= x[n]*c_i[n]
        b+= x[n]**2
        c+= c_i[n]**2

    gamma= a/(np.sqrt(b*c))
    return gamma
```

```python
def fourier(sig,sr):
    X = np.fft.fft(sig)
    N = len(X)
    n = np.arange(N)
    T = N/sr
    freq = n/T 
    return freq[:N//2], np.abs(X)[:N//2]
```

```python
plt.figure(figsize = (12, 6))

plt.subplot(121)
L= fourier(imfm[:,0],fs)
plt.step(1/L[0], L[1], 'b')
L= fourier(imfm[:,1],fs)
plt.step(1/L[0], L[1], 'lightblue')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()

plt.subplot(122)
fxx, pxx =sp.signal.welch(imfm[:,0],fs,nperseg=len(t1))
plt.step(1/fxx,pxx,'b')
plt.axvline(x=3,linestyle='--',color='green')
plt.loglog()
```
