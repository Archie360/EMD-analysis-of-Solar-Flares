```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import emd
```

```python
file=open('vel_fe_029.txt','r')
f=file.readlines()
new_list = []
for line in f:
   
    new_list.append(line[:-1])

    
l2 = []
for a in new_list:
    l = []
    
    l2.append(a.split("\t"))
    
t=[]
wp=[]
for li in l2:
    t.append(li[0])
    wp.append(li[1])
    
t1=[]
for i in t:
    t1.append(float(i))
    
wp1 = []
for j in wp:
    wp1.append(float(j))
```

# EMD

```python
x=np.array(wp1)
```

```python
imf = emd.sift.sift(x)
```

```python
plt.plot(imf)
```

```python
# help(emd.plotting.plot_imfs)
# emd.plotting.plot_imfs(figsize=(6,3))
emd.plotting.plot_imfs(imf)
```

```python
peak_locs, peak_mags = emd.sift.get_padded_extrema(x, pad_width=0, mode='peaks')
trough_locs, trough_mags = emd.sift.get_padded_extrema(x, pad_width=0, mode='troughs')

plt.figure()
plt.plot(x, 'k')
plt.plot(peak_locs, peak_mags, 'o',alpha=0.5)
plt.plot(trough_locs, trough_mags, 'o',alpha=0.5)
plt.xlim(0, 150)
```

```python
proto_imf = x.copy()
# Compute upper and lower envelopes
upper_env = emd.sift.interp_envelope(proto_imf, mode='upper')
lower_env = emd.sift.interp_envelope(proto_imf, mode='lower')

# Compute average envelope
avg_env = (upper_env+lower_env) / 2

plt.figure()
plt.plot(x, 'k')
plt.plot(upper_env)
plt.plot(lower_env)
plt.plot(avg_env)
plt.xlim(0, 150)
# plt.legend(['Signal', 'Upper Env', 'Lower Env', 'Avg Env'])
```
