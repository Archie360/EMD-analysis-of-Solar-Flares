```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
import emd
```

```python
import statistics as stats
```

```python
file=open('inttime_har060_    1848km.txt','r')
f=file.readlines()
new_list = []
for line in f:
   
    new_list.append(line[:-1])

    
l2 = []
for a in new_list:
    l = []
    
    l2.append(a.split("\t"))
    
t=[]
wp=[]
for li in l2:
    t.append(li[0])
    wp.append(li[1])
    
t1=[]
for i in t:
    t1.append(float(i))
    
wp1 = []
for j in wp:
    wp1.append(float(j))
```

```python
dt=[]
for a in range(len(t1)-1):
    dt.append(t1[a+1]-t1[a])
ts=stats.mean(dt)
fs=1/ts
```

```python
x=np.array(wp1)
```

# EMD

```python
x=np.array(wp1)
```

```python
imf = emd.sift.sift(x)
```

```python
# Get the default configuration for a sift
config = emd.sift.get_config('sift')
# Adjust the threshold for accepting an IMF
config['imf_opts/sd_thresh'] = 0.05
config['extrema_opts/method'] = 'rilling'

imf = emd.sift.sift(x)

# emd.plotting.plot_imfs(imf)
```

```python
# Extract the options for get_next_imf - these can be customised here at this point.
imf_opts = config['imf_opts']

# Extract first IMF from the signal
imf1, continue_sift = emd.sift.get_next_imf(x[:, None],           **imf_opts)

# Extract second IMF from the signal with the first IMF removed
imf2, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1,      **imf_opts)

# Extract third IMF from the signal with the first and second IMFs removed
imf3, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2, **imf_opts)

imf4, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2-imf3, **imf_opts)

imf5, continue_sift = emd.sift.get_next_imf(x[:, None]-imf1-imf2-imf3-imf4, **imf_opts)


# The residual is the signal component left after removing the IMFs
residual = x[:, None] - imf1 - imf2 - imf3-imf4-imf5

# Contactenate our IMFs into one array
imfs_manual = np.c_[imf1, imf2, imf3,imf4,imf5, residual]

# Visualise
# emd.plotting.plot_imfs(imfs_manual)
```

```python
# Compute frequency domain features using the normalised-Hilbert transform
IP, IF, IA = emd.spectra.frequency_transform(imf2,fs, 'nht')

# Extract cycle locations
all_cycles = emd.cycles.get_cycle_vector(IP, return_good=False)

# ``all_cycles`` is an array of the same size as the input instantaneous phase.
# Each row contains a vector of itegers indexing the location of successive
# cycles for that IMF.

print('Input IMF shape is - {0}'.format(IP.shape))
print('Input all_cycles shape is - {0}'.format(all_cycles.shape))
```

```python
def list_corrector(l):
    l_=[]
    for i in range(len(l)):
        l_.append(l[i][0])
    return np.array(l_)
```

```python
imf1_=list_corrector(imf1)
imf2_=list_corrector(imf2)
imf3_=list_corrector(imf3)
imf4_=list_corrector(imf4)
```
