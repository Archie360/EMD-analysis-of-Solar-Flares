```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
```

# fourier transform

```python
file=open('inttime_har060_    1848km.txt','r')
f=file.readlines()
new_list = []
for line in f:
   
    new_list.append(line[:-1])

    
l2 = []
for a in new_list:
    l = []
    
    l2.append(a.split("\t"))
    
t=[]
wp=[]
for li in l2:
    t.append(li[0])
    wp.append(li[1])
    
t1=[]
for i in t:
    t1.append(float(i))
    
wp1 = []
for j in wp:
    wp1.append(float(j))
```

```python
plt.grid(b=True, color='w', alpha=0.3, linestyle='-', linewidth=1)
plt.style.use('dark_background')
plt.step(wp1)
```

```python
x=np.array(wp1)
```

```python
N=len(x)
```

```python
freq=np.arange(N)/(N*34)
```

```python
f_n=freq*1000
```

```python
fx= np.fft.fft(x)
```

```python
f_abs=np.abs(fx)
```

```python
psd=f_abs**2
```

```python
plt.grid(b=True, color='w', alpha=0.3, linestyle='-', linewidth=1)
plt.style.use('dark_background')
plt.step(f_n[:N//2],psd[:N//2],'b')
```

```python
t_h = 1/freq
```

```python
plt.grid(b=True, color='w', alpha=0.3, linestyle='-', linewidth=1)
plt.style.use('dark_background')
plt.step(t_h/60,f_abs,where='post',color='w',lw=1)
# plt.xlim(20)
```
